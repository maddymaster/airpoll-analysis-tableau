```python
# loading libraries and reading the data
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

#bangalore
yemalur = pd.read_csv("./Yemalur - Yemalur.csv")
itpl = pd.read_csv("./ITPL - ITPL.csv")
indiranagar = pd.read_csv("./Indiranagar100ft - Indiranagar100ft.csv")
mgroad = pd.read_csv("./mgRoadMetroStation - mgRoadMetroStation.csv")

#delhi
dlfcybercity = pd.read_csv("./DLFCyberCity - DLFCyberCity.csv")
igi = pd.read_csv("./IGI - IGI.csv")
indiagate = pd.read_csv("./IndiaGate - IndiaGate.csv")
shantiniketan = pd.read_csv("./Shantiniketan - Shantiniketan.csv")
```

```python
# Merging the dataframes
# Note that Cust_id is the common column/key, which is provided to the 'on' argument
# how = 'inner' makes sure that only the customer ids present in both dfs are included in the result
bangalore1 = pd.merge(yemalur, itpl, how='outer', on='updatedAt', sort = True)
bangalore2 = pd.merge(indiranagar, mgroad, how='outer', on='updatedAt', sort = True)
bangalore3 = pd.merge(bangalore1, bangalore2, how='outer', on='updatedAt', sort = True)
delhi1 = pd.merge(dlfcybercity, igi, how='outer', on='updatedAt', sort = True)
delhi2 = pd.merge(indiagate, shantiniketan, how='outer', on='updatedAt' , sort = True)
delhi3 = pd.merge(delhi1, delhi2, how='outer', on='updatedAt', sort = True)
```
